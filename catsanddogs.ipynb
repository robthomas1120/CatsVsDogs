```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_path = '/content/drive/MyDrive/Cats and Dogs/cats and dogs/train'
val_path = '/content/drive/MyDrive/Cats and Dogs/cats and dogs/val'
```

```python
train_datagen = ImageDataGenerator(rescale=1./255, rotation_range=20, horizontal_flip=True)
val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_path,
    target_size=(224, 224), # Standard size for most models
    batch_size=32,
    class_mode='binary'    # Use 'binary' for 2 classes (cat vs dog)
)

val_generator = val_datagen.flow_from_directory(
    val_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)
```

```python
base_model = tf.keras.applications.MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
base_model.trainable = False # Freeze the pre-trained layers

model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(1, activation='sigmoid') # Sigmoid for binary classification
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10
)
```

```python
from google.colab import files

# Save the model
model.save('cat_dog_model.keras') # Use .keras instead of .h5

# Download to your laptop
files.download('cat_dog_model.keras')
```
